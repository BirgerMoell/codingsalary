```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import sys  

import numpy as np
import matplotlib

# reload(sys)  
# sys.setdefaultencoding('utf8')
pd.options.display.float_format = '{:,.2f}'.format

sns.set_style("whitegrid")

matplotlib.style.use('seaborn-dark-palette')

font = {'family' : 'normal',
        'weight' : 'normal',
        'size'   : 20}

matplotlib.rc('font', **font)
```

```python
data = pd.read_csv('salary.csv')

data.columns = ['role', 'salary', 'years']

data['role'] = data.role.str.lower()
data['years'] = data.years.apply(pd.to_numeric)

data.head()

```

```python
data.role.describe()
```

```python
fs = (15, 8.27)
fig, ax = plt.subplots(figsize=fs)
sns.distplot(data.salary)
ax.set_xlim([0, 250000])
plt.yticks([], [])

fs = (15, 8.27)
fig, ax = plt.subplots(figsize=fs)
sns.distplot(data.years)
ax.set_xlim([0, 40])
plt.yticks([], [])
plt.show()
```

```python
from scipy import stats

fs = (15, 8.27)
fig, ax = plt.subplots(figsize=fs)

# get coeffs of linear fit
slope, intercept, r_value, p_value, std_err = stats.linregress(data['years'],data['salary'])

# use line_kws to set line label for legend
ax = sns.regplot(x="years", y="salary", data=data, color='#3D9977', 
 line_kws={'label':"salary = {0:.0f} * years + {1:.0f}".format(slope, intercept)})

# plot legend
ax.legend()
plt.title('Linjr regression')

print("R2 = {:.2f}".format(r_value))
print("p = {}".format(p_value))

plt.show()
```

```python
from scipy import stats

# get coeffs of linear fit
slope, intercept, r_value, p_value, std_err = stats.linregress(data['years'], np.log10(data['salary']))

data_logged = data.copy()
data_logged['salary'] = np.log10(data['salary'])

# ax.set_yscale('log')  # set_yscale is a function, not a string

# use line_kws to set line label for legend

fs = (15, 8.27)
fig, ax = plt.subplots(figsize=fs)
sns.regplot("years", "salary", data=data, color='#3D9977', ax=ax, fit_reg=False,)

#ax.set_yscale('log')
plt.plot(np.linspace(0, 40), 10**(intercept + slope*np.linspace(0, 40)), color='#3D9977',
            label="salary = {1:.0f} * {0:.3f}^years".format(10.0**slope, 10.0**intercept))

ax.set_xlim([-0.5, 40])
ax.set_ylim([0, 210000])

ax.legend()

plt.title('Linear regression -> procentuell lnepverkan')

# R2 computed on linear scale
r_value = 1.0 - (data.salary - 10**(intercept + slope*data.years)).std()**2/data.salary.std()**2

print("R2 = {:.2f}".format(r_value))
print("p = {:}".format(p_value))



plt.show()
```

```python
translationer = {'front end utvecklare': 'frontendutvecklare',
'frontend utvecklare': 'frontendutvecklare',
'software utvecklare': 'utvecklare'}

def trans(x):
    if x in translationer:
        return translationer[x]
    return x

data.role = data.role.astype(str).map(lambda x: x.replace('developer', 'utvecklare').strip()).map(trans)
role_salary = data.groupby('role').salary.describe()

```

```python
role_salary = data.groupby('role').salary.describe()
role_salary = role_salary.loc[role_salary['count'] >= 5].sort_values('50%')

fs = (15, 8.27)
fig, ax = plt.subplots(figsize=fs)
ax.bar(range(len(role_salary)), role_salary['50%'], color='#3D9970')
plt.xticks( np.arange(len(role_salary)), role_salary.index, rotation='vertical')
plt.title('Medianln', fontsize=20)
plt.show()
```

```python
role_salary = data.groupby('role').salary.describe()
role_salary = role_salary.loc[role_salary['count'] >= 5].sort_values('mean')

fs = (15, 8.27)
fig, ax = plt.subplots(figsize=fs)
ax.bar(range(len(role_salary)), role_salary['mean'], color='#3D9970')
plt.xticks(np.arange(len(role_salary)), role_salary.index, rotation='vertical')
plt.title('Medelln', fontsize=20)
plt.show()
```
